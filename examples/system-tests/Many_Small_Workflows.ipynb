```python
import aqueduct as aq
from aqueduct.constants.enums import ArtifactType

client = aq.Client(api_key="", aqueduct_address="")
aq.global_config({"lazy": True})
```

```python
snowflake_warehouse = client.resource('snowflake_resource')
hotel_reviews = snowflake_warehouse.sql('SELECT * FROM hotel_reviews;')
```

```python
@aq.op(requirements=[])
def sample_function(df):
    df["positivity"] = 123
    return df

output_df = sample_function(hotel_reviews)
```

```python
# snowflake_warehouse.save(generated_df, table_name="large_hotel_reviews", update_mode="replace")
flow_map = {}

for i in range(100):
    name = "many_small_workflows_"+str(i)
    flow_obj = client.publish_flow(
        name,
        "many workflows at once",
        schedule="*/10 * * * *",
        artifacts=[output_df],
    )
    flow_map[name] = flow_obj
```
